## <font color= 	#483D8B> Linear Regression </font>
---

A straight line graph with equation *y = kx + m*; where *k* is the slope and *m* is the y-intercept.

<font color=	#FF4500>**Independent variable (x):**</font> One that controls the other

<font color=	#FF4500>**Dependent variable (y):**</font> One that is controlled

The sign of slope determines the relation between x and y: 

#### <font color=	#FF4500>1. Strong positive linear pattern:</font>

```python
x= [10, 20, 30, 40, 50, 60, 70]
y= [1, 2.5, 3, 4, 4.5, 6, 7]

plt.scatter(x, y)
plt.show()
```

#### <font color=	#FF4500>2. Positive linear pattern with one deviation (outlier):</font>

```python
x= [10, 20, 30, 40, 50, 60, 70]
y= [1, 2, 3, 2, 4.5, 6, 7]

plt.scatter(x, y)
plt.show()
```

#### <font color=	#FF4500>3. Strong negative linear pattern:</font>

```python
x= [10, 20, 30, 40, 50, 60, 70]
y= [7, 6, 4.5, 4, 3, 2.5, 1]

plt.scatter(x, y)
plt.show()
```

#### <font color=	#FF4500>4. Weak negative linear pattern:</font>

```python
x= [10, 20, 30, 40, 50, 60, 70]
y= [7.5, 4.5, 4.5, 2, 3, 1.5, 1]

plt.scatter(x, y)
plt.show()
```

## <font color= 	#483D8B>Least squares fit</font>
---

<font color=	#FF4500>Residual (or error)</font> is the distance of each point from the best fit line. Least-squares regression used to minimise the sum of squared residuals.


## <font color= 	#483D8B>$R^2$ </font>
---

Gives the percentage variation of *y* that can be explained by the variation in *x*.

For example: 
if r= 0.50, $r^2$= 0.25

Hence there is a 25% chance that *y* is controlled by *x*. Remaining 75% is controlled by other factors.



```python
x= [10, 20, 30, 40, 50, 60, 70]
y= [1, 2.5, 3, 4, 4.5, 6, 7]

slope, intercept, r, p, std_err = stats.linregress(x, y)
res = stats.linregress(x, y)

def myfunc(x):
    
    return slope * x + intercept

#run each x-value through the function

mymodel = list(map(myfunc, x))

#Draw scatter plot and add linear regression to it

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()

print(f"R-squared: {res.rvalue**2:.6f}")
```


## <font color= 	#483D8B> Correlation value </font>
---

Measure of how well x and y follow each other. Value between -1 and 1 (no units).

<font color=	#FF4500>**Positive correlation (+r):**</font> If x increases or decreases; y follows the same trend.

<font color=	#FF4500>**Negative correlation (-r):**</font> If x increases or decreases; y follows the opposite trend.


```python
np.random.seed(100)

#create array of 50 random integers between 0 and 10
var1 = np.random.randint(0, 10, 50)

#create a positively correlated array with some random noise
var2 = var1 + np.random.normal(0, 10, 50)

#calculate the correlation between the two arrays
np.corrcoef(var1, var2)[0,1]
```


## <font color= #483D8B>Hypothesis testing </font>
---

If x and y linearly connected, r must be positive or negative

Testing if r is significantly away from 0

$H_0 : r = 0 $

$H_A : r != 0$

if p<0.05 reject H0: there is linear relationship

if p>0.05 keep H0: there is no linear relationship, can’t use

<font color=	#FF4500>**t-statistic:**</font>

<br> 
<div style="border: 2px solid skyblue">
<br> $$ t=\frac{r \sqrt {n-2}}{\sqrt {1 - r^2}}$$ <p style="text-align:right">  (1) &nbsp;
</div> 



<font color=	#FF4500>**degree of freedom:**</font> 

<br> 
<div style="border: 2px solid skyblue">
<br> $$ df = n-2 $$ <p style="text-align:right">  (2) &nbsp;
</div> 


```python
n=11
r=0.6631
df= float(n)-1
tval = float(r)*np.sqrt(int(n)-2)/np.sqrt(1-float(r*r))  # t-statistic 
pval = stats.t.sf(abs(tval), df)*2  # two-sided pvalue

print('t-statistic = %6.3f pvalue = %6.4f' % (tval, pval))
```

## <font color= 	#483D8B> Prediction </font>
---

Used to get an estimate of *y* as the predicted average for values with a particular value of x.

```python
x= [10, 20, 30, 40, 50, 60, 70]
y= [1, 2.5, 3, 4, 4.5, 6, 7]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    
    return slope * x + intercept

pred_val = myfunc(39)

print(pred_val)
```

## <font color= 	#483D8B> Outlier identification </font>
---

A point that is deviated from rest of the graph and may change the results significantly. 

<br> 
<div style="border: 2px solid skyblue">
<br> $$ s = \sqrt \frac{SSE}{n-2} $$ <p style="text-align:right">  (3) &nbsp;
</div>



Compare reiduals to s.

Any value > 2*s or < -2*s is an <font color=	#FF4500>**outlier**</font>
