## 3. Histograms

**Histograms** are graphical representation of a large set of data and looks like a bar plot. The vertical axis is either frequency (*f*) or relative frequency (*RF*). *RF* equals to the frequency for an observed value of the data divided by the total number of data values in the sample (*n*).

$$
RF = \frac{f}{n}
$$

## 4. Mean, median, mode and standard deviation

**Mean** equals the sum of all values, divided by total number of values (N).

$$
Mean = \frac{\sum Values}{N}
$$

```python
#using Numpy array
x= np.array([1,1,1,2,2,3,4,4,4,4,4])
np.mean(x)
```


**Median** is the value that comes in middle when data is sorted from lowest to highest. For odd number of values, take average of mid-two values.


```python
x= np.array([1,1,1,2,2,3,4,4,4,4,4])
np.median(x)
```

**Mode** is the most repeated number in a dataset.

```python
x= np.array([1,1,1,2,2,3,4,4,4,4,4])    
values,counts = np.unique(x, return_counts= True)
mode= values[np.argmax(counts)]
mode
```

**Standard deviation** is a measure of spread. Tells how far is the real value from the mean.

```python
x= np.array([1,1,1,2,2,3,4,4,4,4,4])
np.std(x)
```

## 5. Box plots

Graphical representation of data set with a box indicating quartiles and median. As median is involved, data has to be arranged in ascending order.
It shows min, LQ, Median, UQ and max.

**Percentile** divides ordered data into hundreths and is not equal to percentage. So $90^{th}$ percentile means 90% of the data are equal to or less than your value and only 10% are equal to or more than your value.

*Median* is the $50^{th}$ percentile.

*Lower Quartile* (LQ) is the $25^{th}$ percentile i.e. median on lower half.

*Upper Quartile* (UQ) is the $75^{th}$ percentile i.e. median on upper half.
