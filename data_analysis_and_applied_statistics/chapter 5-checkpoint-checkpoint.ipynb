## <font color= 	#483D8B> Central Limit Theorem </font>
---
 

If the size (n) of samples is sufficiently large, the distribution of the sample means will be approximately normal.The mean of the sample means will equal the population mean. The standard deviation of the distribution of the sample means, called the <font color=	#FF4500>**standard error (SE)**</font> of the mean, is equal to the population standard deviation ($\sigma$) divided by the square root of the sample size (n).

<br> 
<div style="border: 2px solid royalblue">
<br> $$ SE = \frac {\sigma}{\sqrt n}$$ <p style="text-align:right">  (1) &nbsp;
</div> 





## <font color= 	#483D8B> Emperical rule </font>
---


About 68% of the x values lie between –1σ and +1σ of the mean µ (within one standard deviation of the mean).

About 95% of the x values lie between –2σ and +2σ of the mean µ (within two standard deviations of the mean).

About 99.7% of the x values lie between –3σ and +3σ of the mean µ (within three standard deviations of the mean).



## <font color= 	#483D8B> Confidence Intervals </font>
---


An interval estimate for an unknown population parameter.

For example, sample mean (m) = 2 , population standard deviation ($\sigma$) = 1 , sample size (n) = 100
So, using central limit theorem; 

<br> 
<div style="border: 2px solid royalblue">
<br> $$ SE = \frac {\sigma}{\sqrt n}= \frac {1}{\sqrt 100} = 0.1 $$ <p style="text-align:right">  (2) &nbsp;
</div> 



So, sample standard deviation (SE)  = 0.1

From emperical rule, 95% of values lie between –2σ and +2σ of the mean of sample (m). 

i.e. if m = 2 and SE = 0.1;

95% of the time average is between 1.8 and 2.2 (m \+- SE).

Hence, 1.8-2.2 is the *confidence interval*. 



## <font color= 	#483D8B> Finding z-score (for known standard deviation) </font>
---


Z-score states how unusual the value (x) is.

<br> 
<div style="border: 2px solid royalblue">
<br> $$ z= \frac{x-µ}{\sigma}$$ <p style="text-align:right">  (3) &nbsp;
</div> 



where µ is the mean and $\sigma$ is standard deviation.

For example, to find z-score for 95% confidence level (CL) with µ= 2 and $\sigma$ = 0.1,

calculate $\alpha =1-CL$ :

$\alpha = 1-0.95 = 0.05$

divide by two, as there's a part on each side of the distribution: $ \frac{\alpha}{2} = \frac{0.05}{2}= 0.025$

now use the formula of z-score for 0.025.
