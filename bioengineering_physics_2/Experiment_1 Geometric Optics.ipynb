## <font color="royalblue"> 1 Theory

### <font color="royalblue">1.1 Lens Equations
In geometric optics light is spreading linearly along rays. This assumption allows to treat optical imaging
with the help of trigonometry. Optical imaging means here, that all light beams that emanate from one
point of an object are reunited by an optical system, for example a lens, in one point of the optical image.
The lens as the simplest optical system is characterised by its type, i.e. if it is a converging or diverging
lens, its diameter, or optical aperture and its focal lenght f. The focal length of a converging lens is the
distance of the lens to the focal point. A diverging lens has no focal point, but the backward extensions of
the diverging rays meet in one point, which can be interpreted as a virtual focal point of the diverging lens
(see figure 1). The distance of this virtual focal point to the lens is then again the focal length of the lens.
The optical imaging of a lens can be constructed geometrically by considering the course of three specific
beams that are representatives for all other beams:

    1. Parallel ray: These rays run parallel to the optical axis into the optical system. They become focal
       rays when they pass through a lens, i.e. rays that run through the focal point. 
    
    2. Focal rays: Rays that run through the focal point become parallel rays when they pass through a lens.
    
    3. Center rays: Rays that run through the center of the lens keep their direction.

The parallel, focal and center rays that emanate from one point of an object, are reunited by a converging
lens in one point of the image, so that the object point is mapped onto this image point (see figure 1). This
construction can be done for all points of the object, so that a real image is obtained that can be seen on a
screen. The distance of the object to the lens is the object distance $d_O$ and the distance of the image to the lens
is the image distance $d_I$ . And to the object height $h_O$ corresponds an image height hI . Applying then the
intercept theorem to the similar triangles *FI'I* and *FAM* gives:

<p style="text-align:center"> $ {h_I \over h_O} = {d_I âˆ’ f \over f} $ <p style="text-align:right"> (1) </p>

and applying it to the similar triangles *OMO'* and *IMI'* gives:

<p style="text-align:center"> $ {h_I \over h_O} = {d_I \over d_0} $ <p style="text-align:right"> (2) </p>

Equating both terms gives the lens equation:
<br> <br> 
<div style="border: 2px solid royalblue">
<br> $${1 \over f} = {1 \over d_0} + {1 \over d_I} $$ <p style="text-align:right">  (3) &nbsp;
</div> 


<font color="blue"> *Figure 1: Left: The real image of an object by a converging lens. Right: The virtual image of an object by a diverging lens*

The inverse focal lenght is also called optical refraction power $D = 1/ f$. The unit of the refraction power is
the dioptre, i.e.: $ 1 dpt=1 m^{âˆ’1}$.
The rays that emanate from one point of an object and pass through a diverging lens do not meet again
in one point. But the backward extensions of the rays leaving the lens do, so that a virtual image can be
constructed. This virtual image cannot be seen on a screen, but it can be seen if one looks through the lens.
The converging lens of the eye brings then the diverging rays together in one point, so that a real image is
created on the retina.
Applying the intercept theorem in the case of a diverging lens to the similar triangles *IMI'* and *OMO'*
as well as *IFI'* and *MFA*, leads to the following equations:

<p style="text-align:center"> $ {h_I \over h_O} = {d_I \over d_0} $ and $ {h_I \over h_O} = {d_I âˆ’ f \over f} $ <p style="text-align:right"> (4) </p>

Equating gives:

<p style="text-align:center"> $-{1 \over f} = {1 \over d_0} + {1 \over d_I} $ <p style="text-align:right"> (5) </p>

which corresponds to the lens equation (3) with negative $f$ and $d_I$ . The lens equation (3) can then also be
used for diverging lenses, if one uses the following sign convention:

- The focal length for a converging lens is positive and for a diverging lens negative.
- The image distance is positive if the image is on the opposite side of the lens compared to the object. The image distance is negative, if image and object are on the same side of the lens. The image is real for positive image distances and virtual for negative image distances.
- The image height $h_I$ is positive if the image is upright and negative if the image is upside down.

The real image of a converging lens and the virtual image of a diverging lens has in general a differnt size
compared to the object. The magnification is defined as the relation of image to object height. Equation
(2) gives under consideration of the sign convenction for the magnification *M* the following equation:
<br> <br>
<div style="border: 2px solid royalblue">
<br> $$ M = {h_I \over h_O} = {d_I \over d_0} $$ <p style="text-align:right">  (6) &nbsp;
</div>  

The sign is introduced, because for a converging lens the image height $h_I$ and for a diverging lens the image
distance $d_I$ is negative.
The lens equation (3) allows for an object in object distance $d_O$ from a lens with focal length $f$ to calculate
the image distance $d_I$ , while equation (6) allows to calculate the corresponding magnification. Furthermore
one can calculate the image height with equation (2), so that the following three equations result:

<p style="text-align:center">$ d_0 = {fd_0 \over d_O-f} $, &nbsp;  $|M| = |{-f \over d_O-f}| $ &nbsp; and &nbsp; $h_I={-f \over d_O-f}h_0$ <p style="text-align:right">  (7) </p>

Theses three equations allow to distinguish for a converging lens (f > 0) the following cases:
<br> <br>
<div style="border: 2px solid royalblue">
<br> <p style="text-align:center"> $ âˆž > d_O > 2f â‡’ $ &nbsp; &nbsp; $  f < dI < 2f, |M| < 1 and 0 > h_I > âˆ’h_O, $ &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;  </p>
<p style="text-align:center">  $ i.e. real, reduced and inverted image $ </p>

<p style="text-align:center">  $ d_O = 2f â‡’ $ &nbsp; &nbsp; $ dI = 2f, |M| = 1 and hI = âˆ’hO, $ &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; </p>
<p style="text-align:center"> &nbsp; &nbsp; $ i.e. real, equally sized and inverted image $ </p>

<p style="text-align:center">  $ 2f > dO > f â‡’ $ &nbsp; &nbsp; $ 2f < dI < âˆž, |M| > 1 and âˆ’h_O > h_I > âˆ’âˆž, $ &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; </p>
<p style="text-align:center">  $ i.e. real, magnified and inverted image $ </p>

<p style="text-align:center">  $ d_O = f â‡’ $ &nbsp; &nbsp; $ d_I , |M| and hI divergent, i.e. no image $ &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; </p>

<p style="text-align:center">  $ f > d_O > 0 â‡’ $ &nbsp; &nbsp; $ âˆ’âˆž < d_I < 0, |M| > 1 and âˆž > h_I > h_O, $ &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; </p>
<p style="text-align:center"> &nbsp; &nbsp; $ i.e. virtual, magnified and upright image $ </p> <p style="text-align:right">  (8) &nbsp;
</div>
<br>
For a diverging lens $(f < 0)$ one always has a negative $d_I$ , a magnification $|M|$ smaller than 1 and a positive
image height $h_I$ , i.e. a diverging lens always produces a virtual, reduced and upright image.

### <font color="royalblue"> 1.2 The Bessel method for measuring the focal length of a converging lens

The simplest method for measuring the focal length of a converging lens is by creating a focal point on
a screen and then measuring the distance of the lens to the screen. Besides this direct method it is also
possible to use the lens to create an image of an object on the screen, measure object and image distance
and then calculate f with the lens equation (3).
Another standard method to determine the focal lenght is the Bessel method. Here the distance of
object and screen is kept constant and one creates a reduced and a magnified image on the screen by variing
the position of the lens. The two different object distances for the reduced and magnified image allow to
calculate the focal lenght. In order to derive the corresponding equation one first states that in the case of
an optical imaging the distance s of object and screen can be written as:

<p style="text-align:center"> $ s = d_O + d_I $ <p style="text-align:right"> (9) </p>

Replacing in the lens equation (3) the image distance by $d_I = s âˆ’ d_O$ and solving for $s$, gives:

<p style="text-align:center"> $ s = {d_0^2 \over d_O - f} $ <p style="text-align:right"> (10) </p>

This function is plotted in figure 2 for a focal lenght of $f = 15 cm$. One sees that there are two different object
distances dO1 and dO2 for a given value of s. These object distances can be obtained after transforming (10)
into $d_O^2 âˆ’ sd_O + fs = 0 $ as the quadratic zeros:

<p style="text-align:center"> $ d_{01,2} = {s \over 2} Â± \sqrt{{s^2 \over 4}-fs} $ <p style="text-align:right"> (11) </p>


<font color="blue"> *Figure 2: Left: The two lens positions for the Bessel method. Right: The distance of object and screen as
a function of the object distance, here for a lens with a focal lenght of 15 cm. For values of s above $4f$ one
has for each value of s two possible object distances $d_{O1}$ and $d_{O2}$. Here the two object distances for $s = 70$
cm are sketched.*

For the image distance follows then $d_{I1,2} = s âˆ’ d_{O1,2} = d_{O2},1$.
This means, that for a given distance s between object and screen, there are two positions for a converging
lens where one obtains a sharp image of the object on the screen. Firstly a reduced image for a object distance
$d_{O1}$ (positive sign in (11)) and secondly a magnified image for a object distance $d_{O2}$ (negative sign in (11)).
But this cannot be realised for any distance s of object and screen. Equation (11) gives only a result if the
term beneath the root is positive, i.e. if ${s^2 \over 4} âˆ’ fs > 0$ or $s > 4f$. Besides this minimum distance of object
and screen one has also an optimal range for the values of s, where the values of d_{O2} vary maximally (see
figure 2). The focal length of the lens can be calculated, if one measures the two object distances $d_{O1}$ and $d_{O2}$
as well as the distance s of object and screen. The formula is obtained by first calculating with (11) the
difference of the two object distances:

<p style="text-align:center"> $ d_{01}-d_{02}= 2 \sqrt{{s^2 \over 4}-fs} $ <p style="text-align:right"> (12) </p>

Solving for f gives then the formula for the focal lenght:
<br><br>
<div style="border: 2px solid royalblue">
<br> $$ f = {1 \over 4s} (s^2(d_{01}-d_{02})^2) $$ <p style="text-align:right">  (13) &nbsp;
</div>

### <font color="royalblue"> 1.3 Optical Instruments

#### <font color="royalblue"> 1. The magnifying glass

In order to see the small details of an object one has to bring it as near as possible to the eye. This increases
the visual angle $Î¸$, so that the image on the retina also becomes bigger. But the eye can only accomodate
up to a minimal distance (the near point) which is $N â‰ˆ25 cm$. For smaller distances of the object to the
eye the image becomes out of focus. The maximal visual angle for the minimal distance of 25 cm is then
$Î¸_{max} â‰ˆ tan Î¸ = hO/N$ (see figure 3). With a magnifying glass it is possible to circumvent this restriction
by creating with the lens a virtual, magnified image. A maximal magnification is obtained if one chosses
an object distance, that is a littlebit smaller than the focal length f of the magnifying glass (see (8)). The
virtual image appears then in a big distance $|d_I| > 25 cm$, so that the eye can again accomodate. For $d_O â‰ˆ f$
is the visual angle given by $Î¸'â‰ˆ tan Î¸'= h_O/f$ (see figure 3) and the magnification of the magnifying glass is
given by:
    
<p style="text-align:center"> $ M_{mag.glass} = {Î¸' \over Î¸}= {N \over f} $<p style="text-align:right"> (14) </p>

#### <font color="royalblue"> 2. The microscope
In a microscope two lenses, the objective lens and the eyepiece, are combined. The objective lens first
creates a real, magnified image of the object. The magnification is maximal if the distant of the object to
the objective lens is a littlebit bigger than the focal lenght f1 of the objective lens (see (8)). On the other
hand side as much light from the object as possible should enter the microscope in order to obtain a maximal
resolution. I.e. the object distance dO1 and therefore also the focal length f1 of the objective lens should be
as small as possible. With (6) one obtains for the magnification of the objective lens:

<p style="text-align:center"> $ M_1 = {d_{I1} \over d_{O1}} â‰ˆ {dI1 \over f_1}$ <p style="text-align:right">  (15) </p>

The real image created by the objective lens is the object for the eyepiece, where the eyepice works like a
magnifying glass and creates an again magnified, virtual image. In order to obtain a maximal magnification
of the eyepiece one need a object distance $d_{O2} that is a littlebit smaller than its focal length $f_2$ (see (8)).
The total magnificatoin of the microscope can then be calculated with (15) and (14) as:
<br> <br>
<div style="border: 2px solid royalblue">
<br> $$ M_{micr.} = M_1 M_2 = {d_{I1} \over f_1} {N \over f_2} = {d âˆ’ d_{O2} \over f_1} {N \over f_2} â‰ˆ {(d âˆ’ f2)N \over f_1 f_2} â‰ˆ {d N \over f_1 f_2} $$ <p style="text-align:right">  (16) &nbsp;
</div> 

using, that the distance d of the two lenses (also called tube length) is given by $d = d_{I1} + d_{O2}$ and that the
distance of the lenses is much bigger than the focal length of the eyepiece.

#### <font color="royalblue"> 3. The Keplerian

The Keplerian is like the microscope also a combination of an objective lens and an eyepiece. But contrary
to the microscope is the focal length of the objective lens of the Keplerian very big and the object is in
an infinite distance (compared to the focal length of the objective lens). The rays that emanate from one
point of the object are parallel, so that the objective lens creates a real image in an image distance of the
focal length of the objective lens, i.e. $d_{I1} = f_1$. The real image created by the objective lens is then again
magnified by the eyepiece. The total magnification is $M_K = Î¸'/Î¸$ (see figure 5). For the angle $Î¸$ one has
$Î¸ â‰ˆ tan Î¸ = h_{I1}/f_1$ and for the angle $Î¸'$ follows $0 â‰ˆ tan Î¸' = h_{I1} / f_2 $, so that the total magnification is given by:
<br> <br>
<div style="border: 2px solid royalblue">
<br> $$ M_K = {Î¸' \over Î¸} = âˆ’{f_1 \over f_2} $$ <p style="text-align:right">  (17) &nbsp;
</div> 
    
where the minus sign means that the image is inverted. A maximal magnification is obtained if the objective
lens has a maximal and the eyepiece a minimal focal length.

#### <font color="royalblue">  4. The Galilean
    
In a Galilean one combines a convex objective lens with big focal length and a concave eyepiece with small
focal length (see figure 6). This design has the advantage, that the image is upright and that the telescope
is shorter, since the distance of the lenses is $ f_1 âˆ’ |f_2| $. The total magnification is here:
<br> <br>
<div style="border: 2px solid royalblue">
<br> $$ M_G = {Î¸' \over Î¸} = âˆ’{f_1 \over |f_2|} $$ <p style="text-align:right">  (18) &nbsp;

### <font color="royalblue"> 3) Measurement of the focal length of a converging lens with the Bessel method

<font color="green"> ***In this part of the evaluation of the experiment, two plots are needed. Both plots have an independent calculation for their regression lines even if in this case it is for both a simple polynomial with a slope "a", a constant "b" and no exponential. If needed functions with exponent, without constant, roots or similar things can be used.***
    
#### <font color="royalblue"> Step 2: The Plots and Calculations
    
<font color="green"> ***The first Graph is the plot for the lens equation, therefore a coordinate system with the inverse image distance on the horizontal axis and the inverse object distances on the vertical axis is created. Then all data points of the lens ( that means both the Data of the magnified and the reduced case) are plotted into the coordinate system and calculate one regression line through all data points.*** 


    
##### <font color="deepskyblue"> Plot 1. lens equation

```python
# define the true objective function
def objective(x, a, b):
    return a * x + b

# choose the variables for the x-achse (horizontal) ande the y-achse (vertical)
x, y = D1, D2

# fit the curve to get a regression line
popt, p = curve_fit(objective, x, y)

# summarize the parameter values  
a, b = popt
print('Regression line:'+' %.5f * x + %.5f' % (a, b))

# define a sequence of inputs between the smallest and largest known inputs and define the step size
x_line = np.arange(min(x), max(x), 0.001)

# calculate the output for the range
y_line = objective(x_line, a, b)

# plot the graph with the regression line 
plt.scatter(x, y,label="Values")
plt.plot(x_line, y_line, '--', color='red',label="Regession line") 

# label the graph
plt.title('lens equation')
plt.xlabel('inverse image distance in cm')
plt.ylabel('inverse object distance in cm')
plt.grid() 
plt.legend()
plt.show()
```


<font color="green"> ***The second graph is the plot of the Magnification therefore a coordinate system with ${|h_I| \over h_O}$ on the horizontal axis and  ${d_I \over d_O}$ on the vertical axis is created. Then again all data points of the lens are plotted and one regression line is calculated.***

##### <font color="deepskyblue"> Plot 2. Magnification plot

```python
# define the true objective function
def objective(x, a, b):
    return a * x + b

# choose the variables for the x-achse (horizontal) ande the y-achse (vertical)
x, y = x1, y1

# fit the curve to get a regression line
popt, p = curve_fit(objective, x, y)

# summarize the parameter values  
a, b = popt
print('Regression line:'+' %.5f * x + %.5f' % (a, b))

# define a sequence of inputs between the smallest and largest known inputs and define the step size
x_line = np.arange(min(x), max(x), 0.001)

# calculate the output for the range
y_line = objective(x_line, a, b)

# plot the graph with the regression line 
plt.scatter(x, y,label="Values")
plt.plot(x_line, y_line, '--', color='red',label="Regession line") 

# label the graph
plt.title('Magnification')
plt.xlabel('|hI|/hO in cm')
plt.ylabel('dI/dO in cm')
plt.grid()
plt.legend()
plt.show()
```
