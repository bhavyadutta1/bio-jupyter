## <font color="royalblue"> 1 Theory

### <font color="royalblue"> 1.1 Superposition of Waves
The summation of waves follows the principle of superposition, i.e. the spreading of each wave is not
affected by the other waves and in the places where the waves overlay the single amplitudes add up to a
total amplitude. Phenomena that appear at a fixed place in space due to the superposition of waves are
called interference phenomena.
As an example one can consider the superposition of two one-dimensional waves with the same frequency
ω, the same amplitude A and the same wave number k. The first wave is given by:

<p style="text-align:center"> $ y_1(t, x) = A cos(ωt − kx) $ <p style="text-align:right"> (1) </p>

The second wave should have a phase shift of ϕ relative to the first wave. A phase ϕ corresponds to a path
difference ∆s of the two waves, where a phase of 2π corresponds to a path difference of one wavelength λ.
This means that ∆s and ϕ are connected by:

<p style="text-align:center"> $ ∆s ={λ \over 2π}ϕ={ϕ \over k} $, <p style="text-align:right"> (2) </p>

where the definition of the wave number k = 2π/λ was used. Then the second wave can be written as:

<p style="text-align:center"> $ y_2(t, x) = A cos(ωt − kx + ϕ) = A cos(ωt − kx + k∆s)$ <p style="text-align:right"> (3) </p>

Using the trigonometric sum formula:

<p style="text-align:center"> $ cosα +cosβ = 2 cos({α − β \over 2})cos({α + β \over 2})$ <p style="text-align:right"> (4) </p>

one obtains for the sum of the two waves:

<p style="text-align:center"> $ y_{ges}(t,x)=y_1(t,x)+y_2(t,x) = 2Acos({ϕ \over 2})cos(ωt−kx +{ϕ \over 2})$ <p style="text-align:right"> (5) </p>
<p style="text-align:center"> $ =y_1(t,x)+y2(t,x)=2Acos({k \over 2}∆s)cos(ωt−kx +{k \over 2}∆s)$ <p style="text-align:right"> (6) </p>    

This means that the sum of the two waves is a wave with the same frequency ω, but with a phase ${ϕ \over 2}$ relative
to $y_1(t, x)$ and a new amplitude that is given by:

<p style="text-align:center"> $ A_{total} = 2Acos({ϕ \over 2}) = 2Acos ({k \over 2}∆s) $ <p style="text-align:right"> (7) </p>



<font color="blue"> *Figure 1: In green and blue the two single wave that superpose and in red the resulting wave. The two single waves have a wavelength of $λ = 1$ . Shown are the pointer diagrams and waves at $t = 0$ for constructive
interference, destructive interference and for a phase shift of $ϕ = π/2$ which corresponds to a path difference
of $∆s = {1 \over 4} λ$*

There are now two special cases shown in figure 1:
1) &nbsp; **Constructive Interference** <br> For a phase shift or path difference of

<p style="text-align:center"> $ ϕ = m · 2π $ &nbsp; &nbsp; or &nbsp; &nbsp; $ ∆s = mλ $ <p style="text-align:right">  (8) </p>

with m = 0, 1, 2, 3, . . ., the wave peaks and wave troughs of the two waves coincide, so that the sum is
a wave with twice the amplitude of the single waves. For the values (8) formula (7) gives accordingly
$|A_{total}| = 2A$.

2) &nbsp; **Destructive Interference** <br> For a phase shift or path difference of
<p style="text-align:center"> $ ϕ = (2m-1) · π $ &nbsp; &nbsp; or &nbsp; &nbsp; $ ∆s =  (2m-1){λ \over 2} $ <p style="text-align:right">  (9) </p>

with m = 0, 1, 2, 3, . . ., the wave peaks of one wave lie above the wave troughs of the other wave, so
that the two waves just extinguish each other. For values (9) formula (7) gives accordingly $A_{total} = 0$.


<font color="blue"> *Figure 2: A circular wave with wavelength $λ = 1$, that spreads radially around the origin. In the contour
plot on the right side the red circles are the wave peaks and the black circles are the wave troughs.*

The mathematical description of one-dimensional waves given in (1) can be generalized easily to two-dimensional waves, like for example water waves. For that purpose one goes over to two-dimensional position vectors $\vec{x} = \begin{pmatrix} x \\ y \end{pmatrix}$ and wave vectors $\vec{k} = \begin{pmatrix} k_z \\ k_y \end{pmatrix}$. A two-dimensional plane wave, that moves in the direction of the wave vector $\vec{k}$ is then given by:

<p style="text-align:center"> $z(t, \vec{x}) = A cos(ωt − \vec{k} · \vec{x}) $ <p style="text-align:right">  (10) </p>

Besides these two-dimensional plane waves, there also exist two-dimensional circular waves. Such a circular
wave spreading around the origin can be described by substituting in (1) the coordinate $x$ by the radial
coordinate $ r = \sqrt{x^2+y^2}$:

<p style="text-align:center"> $z(t, \vec{x}) = A cos(ωt − kr) $ <p style="text-align:right">  (11) </p>

see figure 2.
Two circular waves that spread around the points $(−d/2, 0)$ and $(d/2, 0)$, can be written as:

<p style="text-align:center"> $z_{1,2}(t, \vec{x}) = A cos(ωt − kr_{1,2})=A cos(ωt − k \sqrt{(x ± (−{d \over 2})^2+y^2)} $ <p style="text-align:right">  (12) </p>

and the superposition of these two circular waves gives with the sum formula (4):

<p style="text-align:center"> $ z_{total}(t,\vec{x} ) = z_1(t, \vec{x}) + z_2(t, \vec{x}) = 2A cos({k \over 2}(r_2 − r_1))cos(wt-{k \over 2}(r_2 + r_1))$ <p style="text-align:right">  (13) </p>

I.e. the result is a wave with a total amplitude, that depends on the position (x, y):

<p style="text-align:center"> $A_{total}=2A cos({k \over 2}(r_2 − r_1)) = 2A cos({k \over 2}∆s)$ <p style="text-align:right">  (14) </p>

because the path difference of the two waves $∆s = r_2 − r_1$ is position dependent (see figure 3):

<p style="text-align:center"> $ ∆s = r_2 − r_1 = \sqrt{(x-{d \over 2})^2+y^2} - \sqrt{(x+{d \over 2})^2+y^2} $ <p style="text-align:right">  (15) </p>


<font color="blue"> *Figure 3: Left: Two centers for circular waves with distance d. The corresponding path difference of the
two waves in a point $(x, y)$ is $∆s = r_2 − r_1$. Middle: There is constructive interference on the blue curves,
where always two red or two black circles intersect, i.e. two wave peaks or two wave troughs superimpose.
On the other hand side there is destructive interference on the green curves, where always a red and a black
circle intersect, i.e. a wave peak and a wave trough superimpose. Right: The sum of the two circular waves
$z_{total}(t,\vec{x} )$. On the blue curves the amplitude is always $2A$ due to constructive interference and on the green
curves the amplitude is always zero due to destructive interference.*

Positions with the same path difference $∆s$ all lie on a curve. This curve is given as an implicit function by
(15), considering $∆s$ as constant. The corresponding explicit function follows by solving (15) for y:

<p style="text-align:center"> $ y = ± {1 \over 2∆s} \sqrt{4d^2x^2 − 4∆s^2x^2 − d^2∆s^2 + ∆s^4} ≈ ± {d \over ∆s}x $ <p style="text-align:right">  (16) </p>

Inserting now in this function (8), i.e.$ ∆s = mλ$ for m = 0, 1, 2, . . ., gives the places of constructive interference
(the blue curves in figure 3) and inserting (9), i.e. $∆s = (2m − 1) {λ \over 2} $for m = 1, 2, 3, . . . gives the places of
destructive interference (the green curves in figure 3). ∆s is a very small quantity, so that powers in ∆s can
be neglected and it is possible to approximate in (16) the root by a linear function. This means that for
bigger distances to the sources of the two waves the places of constructive and destructive interference lie
approximately on lines. Along the lines of constructive interference one measures then a maximal amplitude
of $2A$ , which means in the case of light waves that along the lines of constructive interference one has maximal
brightness. And on a screen, that intersects these lines one can see in the intersection points bright spots.
On the other hand side is the amlpitude along the lines of destructive interference always zero. And on
the screen one sees in the corresponding intersection points with these lines then dark spots or brightness
minima.

### <font color="royalblue"> 1.2 Diffraction as a phenomenon of interference
If light falls through an opening one can notice, that the light is distracted in directions, that are not allowed
according to the straight beam path assumed in geometric optics. And furthermore this diffraction of light
becomes stronger for smaller openings. This phenomenon can be explained by taking into account, that light
beams are waves, where the behaviour of the waves can then be understood using Huygens’ construction.
Huygens’ construction states, that firstly each point of a wave front is the origin of a circular elementary wave
with the same wavelength and secondly that the new position of the wave front results from the superposition
of all these elementary waves. In figure 4 Huygens’ construction is depicted for two-dimensional waves.




<font color="blue"> *Figure 4: Huygens’ construction for a) a plane wave, b) a circular wave and c) a plane wave that moves
through a single slit. The origins of some elementary waves are shown exemplarily on the wave front. One
can see, that the elementary waves of a plane wave give again a plain wave, while the elementary waves of
a circular wave give again a circular wave. In c) the elementary waves add up to an interference pattern.*

Huygens’ construction can now be applied to the following cases:

<u> **1) Diffraction at a double slit**</u> <br>
If a double slit is illuminated with monochormatic light, each of the two slits can be considered as the source of an elementary wave, so that the distance of the two wave sources is the slit distance d. The superposition of theses two elementary waves produces then an interference pattern on a screen behind the double slit (see figure 5). The distance y of the double slit to the screen is large compared to the distance d of the two slits. This means, that the direction angle $θ$ and the corresponding path difference $∆s$ of the two waves is small. For small values of $∆s$ it is possible to neglect higher powers of ∆s, so that (16) reduces to: 

<p style="text-align:center"> $ {x \over y} = {∆s \over d} $ &thinsp; ⇔ &thinsp; $ tan θ ≈ sin θ = {∆s \over d} $ <p style="text-align:right">  (17) </p>

Inserting now for $∆s$ the values (8) and (9) for constructive and destructive interference gives for the direction
angle $θ_m$ of the m-th maximum and the m-th minimum of the diffraction pattern the following condition:

<br> <br>
<div style="border: 2px solid royalblue">
<br> <p style="text-align:center"> $ Maxima :$&nbsp;&nbsp;&nbsp;$ d sin θ_m =  mλ $  &thinsp;  for m = 1, 2, 3, . . . <p style="text-align:right"> (18) <br> </p>
<p style="text-align:center"> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ Minima :$&nbsp;&nbsp;&nbsp;$  d sin θ_m = (2m - 1){λ\over2} $ &thinsp;   for m = 1, 2, 3, . . . <p style="text-align:right"> (19) <br> </p> </div>
<br>


<font color="blue"> *Figure 5: Left: The double slit and the corresponding diffraction pattern on the screen. Right: The direction angle $θ$ and the path difference $∆s$.*

<u> **2) Diffraction at a diffraction grating**</u> <br>
A diffraction grating has many slits that all have the same distance g, where g is called the grating constant.
Each illuminated slit of the diffraction grating is again the source of an elementary wave, where neighbouring
waves always have the same path difference ∆s in direction θ (see figure 6). The superposition of all these
elementary waves gives a wave, that has always a small amplitude except for a path difference of neighbouring
waves that is ∆s = mλ, i.e. when all the single waves interfere constructively. The condition for a maximum
in the diffraction pattern is then:

<br> <br>
<div style="border: 2px solid royalblue">
<br> <p style="text-align:center"> $ Maxima :$&nbsp;&nbsp;&nbsp;$ g sin θ_m =  mλ $  &thinsp;  for m = 0, 1, 2, . . . <p style="text-align:right"> (20) <br> </p> </div>
<br>


<font color="blue"> *Figure 6: The $N$ illuminated slits of a grating can be seen as the sources of N circular elementary waves. If one considers a point of the screen with coordinates $(x, y)$, one can assume, that the direction angle $θ$ of
this point is approximately the same for all the illuminated slits. This approximation can be made, because
usually the width of the illuminated diffraction grating is much smaller than the distance y of the grating
to the screen (see the picture on the left). But if the direction angle θ to point $(x, y)$ is the same for all the
illuminated slits, then the path difference of neighbouring slits is in this direction always the same and is
given by$ ∆s = g sin θ $ (see the picture on the right). The superposition of many waves under the condition
that neighbouring waves have the same path difference gives only a big total amplitude, if ∆s is a multiple
of the wavelength λ, i.e. if all the waves interfere constructively. This means, that only in directions $θ_m$ with
$∆sm = g sin θm = mλ$ one can see a brightness maximum on the screen.*

<u> **3) Diffraction at a single slit**</u> <br>
If a single slit of width b is illuminated by light, Huygens’ construction states, that each of the infinitely
many points in the spacing of the slit is the source of an elementary wave (see figure 4c). In direction $θ = 0$,
i.e. perpendicular to the single slit there is no path difference for all these elementary waves, so that all
the elementary waves interfere constructively. This means that opposite to the slit there is the maximum of
zeroth order $(m = 0)$.
Just left and right to the maximum of zeroth order is the minimum of first order $(m = 1)$. This minimum
appears in a direction, where for each of the infinitely many elementary waves another elementary wave
exists with a path difference of $λ/2$, so that all elementary waves are extinguished pairwise. This occurs in a
direction where the path difference of the leftmost and rightmost elementary wave have a path difference of
$λ $(see figure 7). The condition for the direction angle of the first minimum is then $b sin θ1 = λ$.
The appearance of the maximum of first order can be explained similarly. The maximum of first order
occurs under a direction angle, where the leftmost and rightmost elementary wave have a path difference of ${3 \over 2} λ$, so that the condition for the direction angle is given by: $ b sin θ_1 = {3 \over 2} λ$. For a direction angle that fulfills this condition one can subdivide all the elementary waves in three sub-groups, where the waves of two of the
three sub-groups have a pairwise path difference of $λ/2$ and are extinguished pairwise, while the maximum
is formed by the waves of the third sub-group (see figure 7).
Similarly one can go on, i.e. the second minimum appears under a direction angle where the path
difference of the outmost elementary waves is $∆s = 2λ$, i.e. for a direction angle that fulfills the condition
$b sin θ2 = 2λ$. In this case the elementary waves can be subdivided in four sub-groups, where the waves in
the first and the second and in the third and the fourth sub-group interfere destructively. In general two
conditions for diffraction at a single slit follow:

<br> <br>
<div style="border: 2px solid royalblue">
<br> <p style="text-align:center"> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ Maxima :$&nbsp;&nbsp;&nbsp;$ b sin θ_m = (2m + 1){λ\over2} $  &thinsp;  for m = 1, 2, 3, . . . <p style="text-align:right"> (21) <br> </p>
<p style="text-align:center"> $ Minima :$&nbsp;&nbsp;&nbsp;$  b sin θ_m = mλ $ &thinsp;   for m = 1, 2, 3, . . . <p style="text-align:right"> (22) <br> </p> </div>
<br>


<font color="blue"> *Figure 7: In direction of the zeroth maximum all elementary waves are in phase, i.e. in this direction the
elementary waves all interfere constructively. In direction of the first minimum exists for each wave in the
left sub-group a partner wave in the right subgroup with a path difference of $λ/2$, so that all elementary
waves interfere pairwise destructively. In direction of the first maximum one can form three sub-groups,
where each elementary wave of the left sub-group has a partner wave in the middle sub-group with a path
difference of $λ/2$, so that the waves of these two sub-groups interfere destructively, and the first maximum is
only constituted by the elementary waves of the third sub-group. In the direction of the second minimum
there are four sub-groups where the elementary waves of the first and the second and the elementary waves
of the third and the fourth sub-group interfere destructively.*

#### <font color="royalblue"> Step 1: The data

```python
# used libraries:
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import math

# 1) Determination of the wavelength by diffraction at a double slit
# Plot 1
m1=[0,1/0.25,2/0.25,3/0.25,4/0.25,5/0.25,6/0.25,7/0.25,8/0.25,9/0.25,10/0.25] # m/d
sr1=[0,0.00245,0.00490,0.00849,0.0109,0.0144,0.0169,0.0193,0.0217,0.0253,0.0277]# distances from the zeroth maximum to the m-th maximum in cm (red laser)

# Plot 2
sg1=[0.00245,0.00490,0.00721,0.00959,0.01208,0.01442,0.01685,0.0193,0.0217,0.0241]# distances from the zeroth maximum to the m-th minimum in cm (green laser)
m2=[(2*1-1)/0.25,(2*2-1)/0.25,(2*3-1)/0.25,(2*4-1)/0.25,(2*5-1)/0.25,(2*6-1)/0.25,(2*7-1)/0.25,(2*8-1)/0.25,(2*9-1)/0.25,(2*10-1)/0.25] # (2m−1)/d

# 2) Determination of the wavelength with a diffraction grating
# Plot1
m3=[0,1/50,2/50,3/50,4/50,5/50] # m/g
sr2=[0,1.85,3.75,5.32,7.53,9.44] # corresponding direction angles θ (red laser) in deg
sg2=[0,1.58,3.03,4.15,5.80,7.35] # corresponding direction angles θ (green laser) in deg


# 5) Measurement of the slit spacing with diffraction
# Plot 1
s3=[0,0.00600,0.00949,0.0166,0.0213,0.0272,0.0320,0.0378,0.0449] 
m4=[0,1*650,2*650,3*650,4*650,5*650,6*650,7*650,8*650] # mλ
```

<font color="green"> ***Now it is possible to calculate and plot the data. It is unnecessary to write down the numbers again, using the variables is enough.*** 
    
### <font color="royalblue"> 1) Determination of the wavelength by diffraction at a double slit 

#### <font color="royalblue"> Step 2: The Plots + Calculations

<font color="green"> ***For the first Plot, a coordinate system with $m/d$ on the horizontal axis and $sinθ_m$ on the vertical axis is created and a regression line through all data points of the maxima for the red laser is shown.*** 
    
##### <font color="deepskyblue"> Plot 1. Maxima

```python
# define the true objective function
def objective(x, a, b):
	return a * x + b

# choose the input and output variables
x, y = m1, sr1

# fit the curve to get a regression line
popt, p = curve_fit(objective, x, y)

# summarize the parameter values  
a, b = popt
print('Regression line:'+' %.5f * x + %.5f' % (a, b))

# define a sequence of inputs between the smallest and largest known inputs and define the step size
x_line = np.arange(min(x), max(x), 0.0001)

# calculate the output for the range
y_line = objective(x_line, a, b)

# plot the graph with the regression line 
plt.scatter(x, y,label="Values")
plt.plot(x_line, y_line, '--', color='red',label="Regession line") 

# label the graph
plt.title('Maxima')
plt.xlabel(' m/d ')
plt.ylabel('sin0')
plt.grid()
plt.legend()
plt.show()
```

<font color="green"> ***Now in the second Plot, a coordinate system with $(2m−1)/d$ on the horizontal axis and $sinθ_m$ on the vertical axis is created and again regression line is shown but this time through all data points of the minima for the green laser.***

##### <font color="deepskyblue">  Plot 2. Minima

```python
# define the true objective function
def objective(x, a, b):
	return a * x + b

# choose the input and output variables
x, y = m2, sg1

# fit the curve to get a regression line
popt, p = curve_fit(objective, x, y)

# summarize the parameter values  
a, b = popt
print('Regression line:'+' %.5f * x + %.5f' % (a, b))

# define a sequence of inputs between the smallest and largest known inputs and define the step size
x_line = np.arange(min(x), max(x), 0.001)

# calculate the output for the range
y_line = objective(x_line, a, b)

# plot the graph with the regression line 
plt.scatter(x, y,label="Values")
plt.plot(x_line, y_line, '--', color='red',label="Regession line") 

# label the graph
plt.title('Minima')
plt.xlabel('2m-1/d')
plt.ylabel('sin0')
plt.grid()
plt.legend()
plt.show()
```
