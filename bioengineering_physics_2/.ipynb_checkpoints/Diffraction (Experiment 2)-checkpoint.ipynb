I) Theory
The purpose of this experiment is to measure the diffraction of light waves. Phenomena that appear at a fixed place in space due to the superposition of waves are called interference phenomena. 
Constructive Interference For a phase shift or path difference of : ϕ = m · 2π or ∆s = mλ
Destructive Interference For a phase shift or path difference of : ϕ = (2m − 1) · π or ∆s = (2m − 1)λ 2  

II) Setup
For the Setup, we have for all different Part experiments different slits and other things for Measurement of the diffraction of light waves. They are double, single slits, diffraction grating slits, a CD and a Lamp. The most important point is the laser. There are two different laser colors (red with a wavelength of 650 nm and green with a wavelength of 532 nm). There is also a Millimetre paper that we must fix on the lane. The lasers (red or green) then shine on the paper and then we mark the maxima or minima on the paper. Then we also write down the wavelength and color of the laser, the Distance between slit and screen, if our markers are minima or maxima and for a few experiments the grating constant. The different slits and other things will be also fixed on the lane for the right Part experiment.


```python
#III) Evaluation
#1) Determination of the wavelength by diffraction at a double slit

#Step 1: The data
m1=[0,1/0.25,2/0.25,3/0.25,4/0.25,5/0.25,6/0.25,7/0.25,8/0.25,9/0.25,10/0.25]
sr1=[0,0.00245,0.00490,0.00849,0.0109,0.0144,0.0169,0.0193,0.0217,0.0253,0.0277]
m2=[(2*1-1)/0.25,(2*2-1)/0.25,(2*3-1)/0.25,(2*4-1)/0.25,(2*5-1)/0.25,(2*6-1)/0.25,(2*7-1)/0.25,(2*8-1)/0.25,(2*9-1)/0.25,(2*10-1)/0.25]
sg1=[0.00245,0.00490,0.00721,0.00959,0.01208,0.01442,0.01685,0.0193,0.0217,0.0241]

#Step 2: The Plots
#Calculation of the corresponding direction angles θ for each minima or maxima:
#θ=cos-1(D/(√d²+D²)) 

#Plot 1
from numpy import arange
from scipy.optimize import curve_fit
from matplotlib import pyplot
import matplotlib.pyplot as plt

# define the true objective function
def objective(x, a, b):
	return a * x + b

# choose the input and output variables
x, y = m1, sr1
...
# curve fit
popt, _ = curve_fit(objective, x, y)
# summarize the parameter values
a, b = popt
print('y = %.5f * x + %.5f' % (a, b))
...
# plot input vs output
pyplot.scatter(x, y)

# define a sequence of inputs between the smallest and largest known inputs
x_line = arange(min(x), max(x), 1)

# calculate the output for the range
y_line = objective(x_line, a, b)

# create a line plot for the mapping function
pyplot.plot(x_line, y_line, '--', color='red')
plt.title('Maxima')
plt.xlabel('m/d')
plt.ylabel('sin0')
pyplot.show()



```
