## <font color="royalblue"> 1 Theory    

### <font color="royalblue"> 1.1 Helmholtz Coils
A steady electric current $I$ that flows through a wire produces a magnetic field. The magnetic field $d\vec{B}$ ($\vec{x}$)
in the point $\vec{x}$ that is produced by a wire element $d\vec{s}$ situated in point $\vec{y}$ has the following properties (see figure 1):

- $d\vec{B}$ ($\vec{x}$) is perpendicular to $d\vec{s}$ and to the unit vector
<p style="text-align:center"> $ \hat{ \vec{r}} = {\vec{r} \over |\vec{r}|} = {\vec{x}-\vec{y} \over |\vec{x}-\vec{y}|} $ <p style="text-align:right"> (1) </p>
- the magnitude of $d\vec{B}$ ($\vec{x}$) is inversely proportional to $r_2$, where $r = |\vec{r}| = |\vec{x}-\vec{y}|$ is the distance of the wire element to the point $\vec{x}$.
- the magnitude of $d\vec{B}$ ($\vec{x}$) is proportional to the current $I$ and the length |$d\vec{s}$| of the wire element.

- the magnitude of $d\vec{B}$ ($\vec{x}$) is proportional to $sin θ$, where $θ$ is the angle between the vectors $d\vec{s}$ and $\hat{ \vec{r}}$.
    


<font color="blue"> *Figure 1: Magnetic field around a wire.*

This can be summarized as:
<p style="text-align:center"> $ d\vec{B} (\vec{x})= - {µ_0 \over 4π} · I · {(\vec{x}-\vec{y})⨉ d\vec{s} \over |\vec{x}-\vec{y}|^3} = - {µ_0 \over 4π} · I · {\hat{ \vec{r}} ⨉ d\vec{s} \over r^2} $ <p style="text-align:right"> (2) </p>

where $µ_0 = 4π · 10^{−7} {Vs \over A_m}$ is the permeability of free space. The total magnetic field at position $\vec{x}$ is then the sum of the contributions of all wire elements, i.e. (2) has to be integrated along the wire. This leads to the
Biot-Savart law:

<p style="text-align:center"> $ \vec{B} (\vec{x}) = - {µ_0 \over 4π} · I · \int{{\hat{ \vec{r}} ⨉ d\vec{s} \over r^2}} = - {µ_0 \over 4π} {\int{\vec{r} ⨉ d\vec{s} \over r^2}} $ <p style="text-align:right"> (3) </p>

The magnetic field of a coil can then be calculated by integrating (3) along a circle with radius $R$ that is
positioned in the x-y-plane and centered at the origin. Near the origin one obtains an almost homogeneous
magnetic field perpendicular to the coil, i.e. a magnetic field in z-direction with:

<p style="text-align:center"> $ B_z(z) = {µ_0IR^2 \over 2(z^2 + R^2)^{3/2}} $ <p style="text-align:right"> (4) </p>

where this formula has two limiting cases, the magnetic field in the origin and the field along the z-axis in
a great distance:

<p style="text-align:center"> $ B_z(0) = {µ_0I \over 2R}$ &thinsp; for $z = 0$ <p style="text-align:right"> (5) </p>
<p style="text-align:center"> $ B_z(z) ≈ {µ_0IR^2 \over 2z^3}$ &thinsp; for $z ⪢ R$  <p style="text-align:right"> (6) </p>

For a coil with $N$ turns the above results have to be multiplied by $N$.
While the magnetic field of a single coil is only approximately homogenous very close to the middle of
the coil, the parallel combination of two identical coils in a distance $d = R$ produces an almost homogenous
field between the coils. This arrangement of coils is called Helmholtz coils and the condition $d = R$ is the
Helmholtz condition. Using the Biot-Savart law (3) to calculate the field between the two Helmholtz coils
leads to an almost constant field in $z$-direction with a field strenght that is given by:

<br> 
<div style="border: 2px solid royalblue">
<br> $$ B_z ≈ {µ_0NI \over (5/4)^{3/2}R} $$ <p style="text-align:right">  (7) &nbsp;
</div>  

where R is the radius of the coils and N is the number of turns of the two coils.

### <font color="royalblue"> 1.2 The Lorentz Force
If an electric charge $q$ moves with a velocity $\vec{v}$ in a magnetic field it is distracted by the Lorentz force that
can be calculated as:
    
<br> 
<div style="border: 2px solid royalblue">
<br> $$ \vec{F_L} = q · (\vec{v} ⨉ \vec{B}) $$ <p style="text-align:right">  (8) &nbsp;
</div> 
    
For a charge that enters a homogenous magnetic field in such a way that its velocity vector is perpendicular to the magnetic field, the Lorentz force causes the charge to move on an circular orbit. An initial velocity of the charge $\vec{v_0} = (v, 0, 0)^T$ and a magnetic field $\vec{B} = (0, 0, B_z)$ result in a Lorentz force that lies in the x-y-plane and acts as the centripetal force for the charge, so that the charge moves on a circular orbit in the x-y-plane. Setting the Lorentz force equalt to the centripetal force leads to:
       
<br> 
<div sty v q ·le="border: 2px solid royalblue">
<br> $$ q · v · B_z = {m · v^2 \over r_o } $$ <p style="text-align:right">  (9) &nbsp;
</div>  

where ro is the radius of the orbit.
A charge can be accelerated up to the velocity $v$ by using a potential difference U. The final kinetic
energy of the charge is then equal to the potential energy of the electrostatic field $ {1 \over 2} mv^2 = qU$ , so that the velocity of the charge is:

<p style="text-align:center"> $ v = \sqrt{{2qU \over m}} $ <p style="text-align:right"> (10) </p>
    
Inserting this result for $v$ in (9) and solving for $r_o$ gives for the radius of the orbit:

<p style="text-align:center"> $ r_0 = {1 \over B_z} \sqrt{{2mU \over q}} $ <p style="text-align:right"> (11) </p>

For electrons that move in an orbit with radius $r_o$ in a cathode ray tube one can then determine with (11)
using $q = e$ and $m = m_e$ the specific charge of the electron as:
         
<br> 
<div sty v q ·le="border: 2px solid royalblue">
<br> $${e \over m_e} = {2U \over B_z^2r_o^2 } $$ <p style="text-align:right">  (12) &nbsp;
</div>

#### <font color="royalblue"> Step 1: The data

```python
# used libraries:
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
import math 
import pandas as pd

# Data Measurement of 𝐼 and 𝑟𝑜 for a fixed acceleration voltage 𝑈 
Data1 = pd.read_excel('Experiment_3_D1.xlsx')
Data1
```

```python
# Data Measurement of U and 𝑟𝑜 for a fixed current 𝐼
Data2 = pd.read_excel('Experiment_3_D2.xlsx')
Data2
```

### <font color="royalblue">  1) Measurement of $I$ and $r_o$ for a fixed acceleration voltage $U$
    
#### <font color="royalblue"> Step 2: The Plots and Calculations


<font color="green"> ***First of all with the formula $ B_z ≈ {µ_0NI \over (5/4)^{3/2}R} $ (formula 7) for each value of $I$ the corresponding value for $B_z$ is calculated and also the square of it because this is what is used later in the plot. Then the values of $2U/r_o^2$ are calculated.***

```python
Ba = (Data1.q * Data1.N * Data1.ro1) / ( ((5/4)**(3/2)) * Data1.R )
Bb = (Data1.q * Data1.N * Data1.ro2) / ( ((5/4)**(3/2)) * Data1.R )
Bc = (Data1.q * Data1.N * Data1.ro3) / ( ((5/4)**(3/2)) * Data1.R )
Bd = (Data1.q * Data1.N * Data1.ro4) / ( ((5/4)**(3/2)) * Data1.R )
B1=Ba**2
B2=Bb**2
B3=Bc**2
B4=Bd**2
U1= (2*Data1.U / Data1.r1**2)
U2= (2*Data1.U / Data1.r2**2)
U3= (2*Data1.U / Data1.r3**2)
U4= (2*Data1.U / Data1.r4**2)
U= U4.tolist() + U3.tolist() + U2.tolist() + U1.tolist()
Bz= B1.tolist() + B2.tolist() + B3.tolist() + B4.tolist()
print('Bz =',Bz)
print('U =',U)
```

##### <font color="deepskyblue"> Plot 1. Measurement of I and r0 for a fixed acceleration voltage U
<font color="green"> ***Now for the Plot a coordinate system with $Bz²$ on the horizontal axis and $2U/r_o² $on the vertical axis is created. Then a regression line through all the Data points is shown.***

```python
fig = plt.figure(figsize=(10, 6))

# define the true objective function
def objective(x, a, b):
    return a * x + b

# choose the variables for the x-achse (horizontal) ande the y-achse (vertical)
x, y = Bz,U

# fit the curve to get a regression line
popt, p = curve_fit(objective, x, y)

# summarize the parameter values  
a, b = popt
print('Regression line:'+' %.5f * x + %.5f' % (a, b))

# define a sequence of inputs between the smallest and largest known inputs and define the step size
x_line = np.arange(min(x), max(x), 0.0000001)

# calculate the output for the range
y_line = objective(x_line, a, b)

# plot the graph with the regression line 
plt.scatter(x, y,label="Values")
plt.plot(x_line, y_line, '--', color='red',label="Regession line") 

# label the graph
plt.title('Measurement of I and r0 for a fixed acceleration voltage U')
plt.xlabel('Bz² in T²')
plt.ylabel('2U/r0² in V/m²')
plt.grid() 
plt.legend()
plt.show()
```
