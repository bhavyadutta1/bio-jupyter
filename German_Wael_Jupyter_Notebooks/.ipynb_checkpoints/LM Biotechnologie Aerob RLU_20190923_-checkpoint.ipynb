###  <font color=E07A5F> 4.5.Vorbereiten der Fermentation </font>

Vor Beginn der Fermentation wird Wasser und Glucose aseptisch dem YPD Medium zugefügt. Das 
Zufügen erfolgt hierbei durch pumpen
 
Jede Fermentation wird mit einem Gesamtvolumen von 500 mL ausgeführt:


<div style="border: 2px solid #8AAA79; padding: 10px;">
400 mL YPD Medium + X mL Glucose + Y mL Wasser = 500 mL Gesamtvolumen
</div>


####  <font color=007BFF> 4.5.1.Einstellung	der	Glucose	Konzentrationen	im	Reaktor </font>
	

Jede Gruppe verwendet eine andere Konzentration. Die benötigten Volumenmengen an Glucoselösung
(X) müssen vorher berechnet werden. Folgende Mengen (g Glucose!) werden benötigt:

0.3 g, 0.6 g, 1.2 g, 2.5 g, 5.0 g, 10 g (für Gruppe 1, 2, 3, 4, 5, 6 respektive)

###  <font color=E07A5F> 4.6.Inokulation </font>


Das Medium wird mit Zellen aus einer Übernachtvorkultur inokuliert. Die Zielkonzentration im Bioreaktor 
beträgt (falls nicht durch die Betreuer anders angekündigt) 5 x 106 Zellen / mL

```python
import dash
from dash import dash_table
import pandas as pd
import dash_core_components as dcc
from dash import html
import plotly.express as px
from dash.dependencies import Input, Output

# Initialize the Dash app
app = dash.Dash(__name__)

# Create a sample DataFrame
data = [
    {
        "t": i,
        "volume": i * 0,
        "od1": i * 0,
        "od2": i * 0,
        "od_mean": i * 0,
        "tcc": i * 0,
        "glucose": i * 0,
        "empty_tube": i * 0,
        "tube_speed": i * 0,
        "remark": i * 0,
    }
    for i in range(10)
]
df = pd.DataFrame(data)

# Define the layout
app.layout = html.Div([
    dash_table.DataTable(
        id='editable-table',
        columns=[
            {"name": ["t"], "id": "time"},
            {"name": ["Probenahme"], "id": "volume"},
            {"name": ["Ethanol"], "id": "od1"},
            {"name": ["Glucose"], "id": "od2"},
            {"name": ["Trocken Masse"], "id": "od3"},
            {"name": ["OD600"], "id": "od_mean"},
            {"name": ["Lebendzellzahl"], "id": "tcc"},
            {"name": ["Gesamt Zellzahl"], "id": "glucose"},
        ],
        data=df.to_dict('records'),
        merge_duplicate_headers=True,
        editable=True,
        export_format='xlsx',
        style_cell={'textAlign': 'left'},
    ),
    
], style={'width': '100%', 'height': '100px'})



# Run the app
if __name__ == '__main__':
    app.run_server(jupyter_mode="inline", port=8074, debug=False)
```
