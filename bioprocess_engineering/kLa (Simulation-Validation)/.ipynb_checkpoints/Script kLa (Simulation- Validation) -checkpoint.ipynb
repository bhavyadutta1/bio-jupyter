To understand the effect of bioreactor design, power input to agitator and other variables on the volumetric mass transfer coefficient for oxygen values (kLa).

## <font color=CEA0AE> 2. Background </font>

The volumetric mass-transfer coefficient for oxygen (kLa) describes the efficiency with which oxygen can be delivered to a bioreactor. For optimum growth of microorganism or cell cultures in a bioreactor, the mass transfer rate of oxygen to the liquid broth must equal or exceed the rate at which growing cells take up that oxygen. Hence measurements of kLa provide important information about a bioprocess or bioreactor.

Measurements of kLa are made to evaluate new reactor designs, new gas sparging equipment, and/or operating conditions. Among others, aeration rates, power input (power-per-volume) and bioreactor design are especially important for the energetic optimization and scaling up of aerobic fermenters. Different studies have developed different empirical formulas for predicting the kLa.

Using an empirical model in the Simulink-script (KLA_V1), prediction of pO2 values are done in accordance to changes of the following variables: ”Types and numbers of impeller, impeller diameter, vessel inner diameter, working volume of the bioreactor, agitation speed, gassing rate and process temperature”.

## <font color=CEA0AE> 5. Material and methods </font>

For this exercise, simulation experiments are done using a Script in Simulink (KLA_V1). Real experimental data for model validation will be obtained in an online experiment. The volumetric mass-transfer coefficient (kLa) is calculated using the “**Dynamic Gassing-out Method**”.

### <font color=9D6A89> 5.1 Introduction to the “Dynamic Gassing-out Method” </font>

In practice, when using the APPLIKON bioreactor (Exercise-2), the changes of dissolved oxygen concentration (pO2= CL) in water as a model culture medium is measured by sparging either air or nitrogen into the liquid. Firstly, by sparging of nitrogen gas through the culture medium, the pO2 values are gradually decreased to zero. The supply of nitrogen gas is then replaced by air and oxygen dissolves back into solution replacing nitrogen gas. 
Using the Simulink script (KLA_V1), only the second phase where the pO2 increases are predicted.

For the determination of the kLa value multiple readings of the pO2 are taken in the second phase.

The kLa can be deduced based on the following formula by plotting ln(100%/(100%-pO2)) vs. time (Equation 1). 

$$ln*\displaystyle \Bigg(\frac{100\%}{100\%- p_{O_{2}L}}\Bigg)= K_{L}a * t $$
<div style="direction:rtl"> (1) </div>

The kLa coefficient can be determined from the slope of the ln f(CL) vs. time graph (Figure 1).

#### 5.2.2 Validation of the model with real experimental data (Exercise -2)

Using the APPLIKON reactor, an online experiment will be performed.

**5.2.2.1. Experimental set‐up**

The dynamic gassing-out method is applied in water as a medium. The bioreactor is equipped with a temperature and dissolved oxygen (pO2) probe, which have been polarized overnight.

For calibration of the pO2 probe, the bioreactor has to be brought to a “steady-state”, which means the temperature and aeration rate should be maintained for a sufficient period of time (30 minutes) until steady readings of the two parameters are achieved. The pO2 probe has been calibrated at sat- uration point to 100% using air and 0% using nitrogen gas. The reactor contains 2 6-bladed Rushton- Turbine. The calculation of the aeration is done in vvm as shown in equation 2:


$$ Aeration\;rate= 1*\displaystyle\frac{Volume_{gas}}{Volume_{liquid}*t_{min}}=\frac{L}{L*min}=v/vm$$
<div style="direction:rtl"> (2) </div>

**5.2.2.2. Tasks**

The combination of three different process parameters will be tested, namely the agitation speed, the aeration rate and the medium volume. Each group will be assigned for one combination of these parameters (e.g. Group 1: 100 rpm, 0.5 VVM and 500 ml medium).

- With the given values of the different parameters, use the Simulink model to predict the KLa value (adjust all the other parameters in the Simulink to the real values for the used bioreactor). 

- During the online KLa experiment, the same values of the different parameters will be used to operate the APPLIKON bioreactor.

 -   Set the aeration rate, the agitation speed and the volume of the medium to the set points assigned to your group.
