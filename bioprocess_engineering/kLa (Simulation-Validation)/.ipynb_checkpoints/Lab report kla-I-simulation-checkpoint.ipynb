## <font color=B6A6CA> Material and Methods </font>
---
Experiments have been performed according to the lab course protocols. Please state the appropriate protocols including versions.

A simulink model (kLa_V1) is used for the simulation experiment. The model is based on an empirical formula for kLa and power input calculations. The model correlates the variables included in the empirical formula with the time dependent development of pO2 in bioreactor. 


kLa values are then determined using the dynamic method in water by plotting the oxygen saturation (pO2) as a function of time and calculated based on linear regression in the linear range using the following formula:

$$ln*\displaystyle \Bigg(\frac{100\%}{100\%- p_{O_{2}L}}\Bigg)= K_{L}a * t $$
