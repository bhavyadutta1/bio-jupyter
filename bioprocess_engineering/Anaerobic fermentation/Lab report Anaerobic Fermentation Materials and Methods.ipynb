## <font color=E07A5F> Materials and Methods </font>


### <font color=8AAA79> Equipment </font>

Blusens Yieldmaster system, see tutorial: https://www.bluesens.com/products/analysis-systems/yieldmaster

### <font color=8AAA79> Media </font>

1. YPD Medium w/o glucose (final concentrations):
 -  2% Peptone
 -  1% Yeast extract
 -  sterilize by autoclaving.
2. Glucose:
 -  20% glucose in water
 -  Prepare and sterilize by autoclaving

**Remark: a standard YPD medium contains 2% glucose.**

## <font color=285238> Media preparation and preparation of the bioreactors (day 1) </font>

### <font color=8AAA79> Media </font>

Several media for fermentation and analytics, as well as agar plates for the dilution series have to be prepared. Please follow the instructions of the supervisors.

### <font color=8AAA79> Bioreactors </font>

For measuring gas production, BlueSens fermenters are used.

Each fermentation is performed with a total volume of 500 ml YPD medium. 
For autoclavation, prepare the fermenter with the solid components to yield 500 ml YPD medium. Dissolve the solid components with 400 ml (!) water. 
Close all openings with caps and autoclave.


*Please note: all (blue) caps or gas sensors mounted to the large threads (GL 45) must be additional sealed with sealing rings, otherwise the fermenter is not gas-tight and gas flow measurement is not possible!*

## <font color=285238> Fermentation (day 2) </font>

### <font color=8AAA79> Glucose addition </font>

The initial amount of glucose is added according to Table 2. Addition is performed in a sterile manner using a 20% glucose solution using peristaltic pumps.



| Fermenter n° | Glucose (g) |
|--------------|-------------|
| 1            | 0 g         |
| 2            | 0.6 g       |
| 3            | 1.2 g       |
| 4            | 2.5 g       |
| 5            | 5.0 g       |
| 6            | 10.0 g      |


*Table 2: Initial amount of glucose to be added.*

### <font color=8AAA79> Adjusting media volume </font>

In the next step, sterile demineralized water is added using peristaltic pumps in a sterile manner to reach a final volume of 500 ml (400 ml medium (day 1) + X ml Glucose + Y ml water = 500 ml).


### <font color=8AAA79> Inoculation </font>

Inoculation is performed with 20 ml Saccharomyces cerevisiae suspension in a sterile manner using peristaltic pumps. Suspension is prepared using a fresh block of bakers yeast suspended in 200 ml YPD medium.

### <font color=8AAA79> Start fermentation </font>

Immediately after inoculation, fermenters are transferred into the water bath (30°C) and connected to the CO2, EtOH and gas flow sensors. Make sure that you do not forget the rings!

### <font color=8AAA79> Sampling </font>

Remember: sampling is performed by EACH group. Prepare a table, where you can collect all data in a user-readable form during the experiment!

Sample volume: 1.5 ml; for samples, where also biomass is assessed: 5 ml. 
Sampling intervals: 30 minutes.

Analysis (see methods script of applied microbiology lab course!):

 -  All samples: OD, total cell counts
 -  CFU: will be announced by supervisors, in general t0, after spiking and at the end of the experiment.
 -  Biomass: t0 and tend

### <font color=8AAA79> Glucose spiking </font>

Spiking with 2.5 g is performed, after gas production reaches a plateau. 
In general, this is reached after 2 hours (groups with 0 g and 0.6 g glucose) or after three hours (groups with 1.2 g and 2.5 g glucose). 
Spiking is performed by pipetting in the appropriate amount of glucose 20%.
